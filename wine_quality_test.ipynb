# 3. Machine Learning Algorithm Implementation

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

```python
df.corr()
```

```python
"""
   Removing elements with lower or negative correlation 
"""
df=df.drop(['chlorides','free sulfur dioxide','pH','residual sugar','density','total sulfur dioxide',
           'volatile acidity'], axis=1)
```

```python
df.corr()
```

```python
x=df.drop('quality', axis=1)
y=df['quality']
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.metrics import confusion_matrix as cm,classification_report as cr
from sklearn.metrics import accuracy_score
```

### MODEL 1(n_neighbors = 5)

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
predict = knn.predict(X_test)
```

```python
print(cm(y_test,predict))
print("*************************************************")
print(cr(y_test,predict))
```

```python
print("Accuracy of model is: ", round(accuracy_score(y_test,predict),2))
print("Accuracy Percentage is: ",round(accuracy_score(y_test,predict)*100,2),"%" )
```

### MODEL 2(n_neighbors = 6)

```python
knn = KNeighborsClassifier(n_neighbors=6)
```

```python
knn.fit(X_train,y_train)
predict = knn.predict(X_test)
```

```python
print(cm(y_test,predict))
print("*************************************************")
print(cr(y_test,predict))
```

```python
print("Accuracy of model is: ", round(accuracy_score(y_test,predict),2))
print("Accuracy Percentage is: ",round(accuracy_score(y_test,predict)*100,2),"%" )
```
